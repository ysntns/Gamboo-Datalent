# Health Data Analysis Report

This paper presents a K-Means clustering analysis of a healthcare dataset. The aim is to uncover hidden patterns in the dataset and transform this information into business strategies.

## Methodology
K-Means algorithm was used for clustering analysis. The model was tested for different values of `K` and the optimal number of clusters was determined by the elbow method.
